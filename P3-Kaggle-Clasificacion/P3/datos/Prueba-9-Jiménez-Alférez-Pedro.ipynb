```python

df_train.Asientos = df_train.Asientos.fillna(5.0)
df_train.Año = df_train.Año.fillna(2010)
df_train.Ciudad = df_train.Ciudad.fillna('C')
df_train.Combustible = df_train.Combustible.fillna('Diesel')
df_train.Consumo = df_train.Consumo.fillna('19.27 kmpl')
df_train.Descuento = df_train.Descuento.fillna(7.85)
df_train.Kilometros = df_train.Kilometros.fillna(62000)
df_train.Mano = df_train.Mano.fillna('First')
df_train.Motor_CC = df_train.Motor_CC.fillna('936 CC')
df_train.Nombre = df_train.Nombre.fillna('Mercedes-Benz New C-Class C 220 CDI Avantgarde')
df_train.Potencia = df_train.Potencia.fillna('170 bhp')
df_train.Precio_cat = df_train.Precio_cat.fillna(3)
df_train.Tipo_marchas = df_train.Tipo_marchas.fillna('Manual')








df_test = df_test.fillna(12.23)
```

```python
from sklearn.ensemble import RandomForestClassifier

from sklearn import metrics
from sklearn.model_selection import cross_val_score
```

```python
labelAsientos = LabelEncoder().fit(df_asientos.Asientos)
labelAO = LabelEncoder().fit(df_ao.Año)
labelCiudad = LabelEncoder().fit(df_ciudad.Ciudad)
labelCombustible = LabelEncoder().fit(df_combustible.Combustible)
labelConsumo = LabelEncoder().fit(df_consumo.Consumo)
labelDescuento = LabelEncoder().fit(df_descuento.Descuento)
labelKilometros = LabelEncoder().fit(df_kilometros.Kilometros)
labelMano = LabelEncoder().fit(df_mano.Mano)
labelMotorCC = LabelEncoder().fit(df_motor_cc.Motor_CC)
labelNombre = LabelEncoder().fit(df_nombre.Nombre)
labelPotencia = LabelEncoder().fit(df_potencia.Potencia)
labelPrecioCat = LabelEncoder().fit(df_precio_cat.Precio_cat)
labelTipoMarchas = LabelEncoder().fit(df_tipo_marchas.Tipo_marchas)






```

```python
df_train.Asientos = labelAsientos.transform(df_train_orig.Asientos)
df_train.Año = labelAO.transform(df_train_orig.Año)
df_train.Ciudad = labelCiudad.transform(df_train_orig.Ciudad)
df_train.Combustible = labelCombustible.transform(df_train_orig.Combustible)
df_train.Consumo = labelConsumo.transform(df_train_orig.Consumo)
df_train.Descuento = labelDescuento.transform(df_train_orig.Descuento)
df_train.Kilometros = labelKilometros.transform(df_train_orig.Kilometros)
df_train.Mano = labelMano.transform(df_train_orig.Mano)
df_train.Motor_CC = labelMotorCC.transform(df_train_orig.Motor_CC)
df_train.Nombre = labelNombre.transform(df_train_orig.Nombre)
df_train.Potencia = labelPotencia.transform(df_train_orig.Potencia)
df_train.Precio_cat = labelPrecioCat.transform(df_train_orig.Precio_cat)
df_train.Tipo_marchas = labelTipoMarchas.transform(df_train_orig.Tipo_marchas)
df_test.Asientos = labelAsientos.transform(df_test_orig.Asientos)
df_test.Año = labelAO.transform(df_test_orig.Año)
df_test.Ciudad = labelCiudad.transform(df_test_orig.Ciudad)
df_test.Combustible = labelCombustible.transform(df_test_orig.Combustible)
df_test.Consumo = labelConsumo.transform(df_test_orig.Consumo)
df_test.Descuento = labelDescuento.transform(df_test_orig.Descuento)
df_test.Kilometros = labelKilometros.transform(df_test_orig.Kilometros)
df_test.Mano = labelMano.transform(df_test_orig.Mano)
df_test.Motor_CC = labelMotorCC.transform(df_test_orig.Motor_CC)
df_test.Nombre = labelNombre.transform(df_test_orig.Nombre)
df_test.Potencia = labelPotencia.transform(df_test_orig.Potencia)
df_test.Tipo_marchas = labelTipoMarchas.transform(df_test_orig.Tipo_marchas)
```
